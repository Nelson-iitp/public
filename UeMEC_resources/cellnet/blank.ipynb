```python
import torch as tt
import torch.nn as nn
import torch.optim as optim
import numpy as np
from numpy.random import default_rng
import matplotlib.pyplot as plt 
rng = default_rng(12)
import gym
from stable_baselines3 import PPO
import UEMEC.game as game
import UEMEC.core as core
```

```python
n_bsv, n_uav, n_iot = 2, 5, 10
params = core.PARAMS(n_bsv, n_uav, n_iot)
env = core.UeMEC(params=params, seed=None, logging="", frozen=False, max_episode_steps=1)
env.reset
```

```python
# Parallel environments
#env = make_vec_env("CartPole-v1", n_envs=4)
model = PPO("MlpPolicy", env, verbose=1)
model.learn(total_timesteps=25000)
model.save("ppo.pie")

```

```python
del model # remove to demonstrate saving and loading
model = PPO.load("ppo.pie")
def predict_agent(state):
    #return rng.uniform(-1, 1, env.A.shape)
    action, _states = model.predict(state)
    print('Action:', action)
    return action
```

```python
env.print=print
gamer = game.play(env, predict_agent)
```
